```python
# ACD_AssignmentScipy&Matplotlib

# Problem Statement

# Matplotlib

# data to use:
# url=https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv

import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt

titanic = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
```

```python
titanic.info()
```

```python
# Charts to plot:
# 1. Create a pie chart presenting the male/female proportion

count=titanic['sex'].value_counts()
colors = ['red', 'green']
plt.pie(count.values.tolist(), labels=count.index.values.tolist(), colors=colors,startangle=90,autopct='%.1f%%')
plt.title('Male/Female proportion')
```

```python
# 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

titanic = titanic.dropna(subset=['sex'])
mapping = {'male' : 'red', 'female' : 'green'}
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5, c=titanic['sex'].map(mapping))
plt.show()
```
